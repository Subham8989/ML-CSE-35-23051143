```python
import pandas as pd
```

```python
df = pd.read_csv("data/data1.csv", index_col="serial")
```

```python
df
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
le = LabelEncoder()
df["decision_encoded"] = le.fit_transform(df["decision"])
```

```python
X = df[["income", "lawn size"]]
y = df["decision_encoded"]

cart = DecisionTreeClassifier(
    criterion="gini",
    max_depth=3,
    random_state=42
)

cart.fit(X, y)
```

```python
plt.figure(figsize=(14,6))
plot_tree(
    cart,
    feature_names=["income", "lawn_size"],
    class_names=le.classes_,
    filled=True
)
plt.show()
```

```python
sample = [[70, 21]]
prediction = cart.predict(sample)

print("Prediction:", le.inverse_transform(prediction)[0])
```
